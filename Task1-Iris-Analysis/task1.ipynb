## 1. Introduction and Problem Statement

The Iris dataset is one of the most famous datasets in data science.
It contains measurements of 150 iris flowers from 3 different species.
The goal of this task is to explore, summarize and visualize the dataset
to understand the relationships between different features.

## 2. Dataset Understanding and Description

The Iris dataset contains 150 rows and 5 columns. The 4 input features
are sepal length, sepal width, petal length and petal width — all
measured in centimeters. The target variable is species which has
3 classes: Setosa, Versicolor and Virginica.

This dataset is built into the seaborn library so no external file
download is required.

```python


# Importing required libraries
import pandas as pd                  # For data loading and manipulation
import matplotlib.pyplot as plt      # For creating visualizations
import seaborn as sns                # For advanced visualizations

print("All libraries imported successfully!")
```

## 3. Data Loading and Inspection

We load the dataset using seaborn and inspect its structure using
shape, columns, head() and describe() to understand what we are
working with before creating visualizations.

```python
# ================================================
# Loading the Iris Dataset
# We use seaborn's built-in dataset loader
# No need to download any CSV file!
# ================================================

df = sns.load_dataset('iris')        # Load iris dataset directly

# Display basic structure of the dataset
print("Shape of dataset (rows, columns):", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
# ================================================
# Dataset Summary Statistics
# Understanding the spread and range of values
# ================================================

print("Basic Statistics of the Dataset:")
print(df.describe())

# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
print("\nNo missing values found — dataset is clean and ready!")

# Check species distribution
print("\nNumber of samples per species:")
print(df['species'].value_counts())
```

## 4. Exploratory Data Analysis (EDA)

We create three types of visualizations to understand the data better.

A scatter plot shows relationships between variables and helps us
see if species can be separated based on measurements.

A histogram shows the distribution of sepal length values and helps
us understand how frequently each value appears in the dataset.

A box plot shows the spread of petal length values across different
species and helps us detect any outliers in the data.

```python
# ================================================
# Visualization 1: Scatter Plot
# Purpose: Analyze relationships between variables
# We plot sepal length vs petal length
# colored by species to see separation
# ================================================

plt.figure(figsize=(8, 5))          # Set figure size

# Create scatter plot with species as color
sns.scatterplot(
    data=df,
    x='sepal_length',               # X axis: sepal length
    y='petal_length',               # Y axis: petal length
    hue='species',                  # Color by species
    palette='Set1'                  # Color palette
)

plt.title('Sepal Length vs Petal Length by Species')  # Title
plt.xlabel('Sepal Length (cm)')     # X axis label
plt.ylabel('Petal Length (cm)')     # Y axis label
plt.legend(title='Species')         # Add legend
plt.tight_layout()                  # Adjust layout
plt.savefig('scatter_plot1.png')     # Save the plot
plt.show()                          # Display the plot

print("Scatter plot saved successfully!")
```

```python
# ================================================
# Visualization 2: Histogram
# Purpose: Examine distribution of sepal length
# Shows how frequently each value appears
# ================================================

plt.figure(figsize=(8, 5))          # Set figure size

# Create histogram for sepal length
df['sepal_length'].hist(
    bins=20,                        # Number of bins
    color='steelblue',              # Bar color
    edgecolor='black'               # Bar border color
)

plt.title('Distribution of Sepal Length')  # Title
plt.xlabel('Sepal Length (cm)')     # X axis label
plt.ylabel('Frequency')             # Y axis label
plt.tight_layout()                  # Adjust layout
plt.savefig('histogram1.png')        # Save the plot
plt.show()                          # Display the plot

print("Histogram saved successfully!")
```

```python
# ================================================
# Visualization 3: Box Plot
# Purpose: Detect outliers and spread of values
# Shows median, quartiles and outliers
# for petal length across each species
# ================================================

plt.figure(figsize=(8, 5))          # Set figure size

# Create box plot grouped by species
sns.boxplot(
    data=df,
    x='species',                    # X axis: species groups
    y='petal_length',               # Y axis: petal length values
    hue='species',                  # Color by species
    palette='pastel',               # Color palette
    legend=False                    # Hide legend
)

plt.title('Petal Length Distribution by Species')  # Title
plt.xlabel('Species')               # X axis label
plt.ylabel('Petal Length (cm)')     # Y axis label
plt.tight_layout()                  # Adjust layout
plt.savefig('boxplot1.png')          # Save the plot
plt.show()                          # Display the plot

print("Box plot saved successfully!")
```

## 7. Conclusion and Key Insights

The Exploratory Data Analysis of the Iris dataset revealed several
important patterns and insights.

The scatter plot shows that Setosa is clearly separated from the
other two species based on petal length and sepal length. Versicolor
and Virginica overlap slightly making them harder to distinguish
from each other using sepal measurements alone.

The histogram shows that sepal length follows a roughly normal
distribution with most values falling between 5 and 7 centimeters.
This suggests that sepal length is fairly consistent across the
entire dataset.

The box plot reveals that Virginica has the longest petals while
Setosa has the shortest petals. Versicolor falls in between the
two. A few outliers were detected in the Versicolor species which
may require further investigation.

Overall the Iris dataset shows that petal measurements are more
useful than sepal measurements for distinguishing between the three
species. Setosa is the most distinct species while Versicolor and
Virginica share similar characteristics. This makes petal length
and petal width the most important features for any future
classification model trained on this dataset.
